```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("/Users/iTenki/Documents/GitHub/Kaggle/Digit-Recognizer/train.csv")
test = pd.read_csv("/Users/iTenki/Documents/GitHub/Kaggle/Digit-Recognizer/test.csv")
```

```python
params={
'booster':'gbtree',
'objective': 'multi:softmax',
'num_class':10,
#'gamma':0.1,
#'max_depth':12,
#'lambda':2, 
#'subsample':0.7, 
#'colsample_bytree':0.7, 
#'min_child_weight':3, 
#'eta': 0.007,
#'seed':1000,
#'eval_metric': 'auc'
}
```

```python
plst = list(params.items())
num_rounds = 5000
```

```python
train_set,validation_set = train_test_split(train, test_size = 0.3,random_state=1)

y = train_set.label
X = train_set.drop(['label'],axis=1)
val_y = validation_set.label
val_X = validation_set.drop(['label'],axis=1)
```

```python
xgb_val = xgb.DMatrix(val_X,label=val_y)
xgb_train = xgb.DMatrix(X, label=y)
xgb_test = xgb.DMatrix(test)
```

```python
watchlist = [(xgb_train, 'train'),(xgb_val, 'val')]
```

```python
model = xgb.train(plst, xgb_train, num_rounds, watchlist,early_stopping_rounds=100)
model.save_model('./xgboost.model')
```

```python
preds = model.predict(xgb_test,ntree_limit=model.best_ntree_limit)
```

```python
np.savetxt('submission.csv',np.c_[range(1,len(test)+1),preds],delimiter=',',header='ImageId,Label',comments='',fmt='%d')
```
